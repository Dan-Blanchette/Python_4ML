## Dataset

    - Number of data samples: 569
    
    - Each data sample has 30 numeric features/attributes. The first 10 features were directly calculated using mean feautues of all nuclei in an image
    
    - Class labels
        : 212 Malignant (0)
        : 357 Benign (1)
        
    https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_breast_cancer.html#sklearn.datasets.load_breast_cancer

```python
import sklearn.datasets as ds
import numpy as np
import math
```

```python
breast_ds = ds.load_breast_cancer()
print('Data fields in breast_ds: \n', dir(breast_ds))

print('\n Dataset description:\n', breast_ds['DESCR'])
```

```python
# we are going to use the first 10 features in this assignment.
ftrs = breast_ds.data[:, :10]
tgts = breast_ds.target

print('Size of feature vectors: ', ftrs.shape)
print('Size of targets: ', tgts.shape)

print('Target labels:\n', tgts)
```

### Task 1: Explore the dataset.  30 points <a id = "section1"/>
    - count the number of data samples for each category (0 or 1 in tgts). 
    - calculate the the mean feature vectors.
    - plot the mean feature vectors.

```python
# count and print out the number of malignant samples (0)
mal = (tgts == 0)
mal_cnt = (tgts==0).sum()
print("There are", mal_cnt, "malignant samples present")

# count and print out the number of benign samples (1)
ben = (tgts == 1)
ben_cnt = (tgts== 1).sum()
print("There are", ben_cnt, "benign samples present")
```

```python
# calculate the mean of all feature vectors

# take all the data column by column and find the mean value
all_feat_mean = np.mean(ftrs, axis=0)
print("The mean of all the feature vectors is:",all_feat_mean, "\n")

# create empty vectors from tgts and add an empty vector of 10 rows
# to match the dimensions of the feature array.
ben_ftrs = np.empty([0, 10])
mal_ftrs = np.empty([0, 10])

# print(ben_ftrs) # check dimensions
for i, rows in enumerate(ftrs):
    # if the target 
    if tgts[i]:
        ben_ftrs = np.vstack((ben_ftrs, rows))
    else:
        mal_ftrs = np.vstack((mal_ftrs, rows))
        
# mean of benign features vectors

# calculate the mean value of the benign data
# column by column
ben_mean = np.mean(ben_ftrs, axis=0)
print("The benign mean is: ", ben_mean,"\n")

# mean of malignant feature vectors
# calculate the mean value of the malignant data
# column by column
mal_mean = np.mean(mal_ftrs, axis=0)
print("The malignant mean is: ", mal_mean, "\n")
```

```python
# show the mean feature vectors using matplotlib 
from matplotlib import pyplot as plt
plt.plot(all_feat_mean, 'go', label='all feature mean')

#show the mean feature vector of benign samples using blue dots
plt.plot(ben_mean, 'bo', label='benign mean')



#show the mean feature vector of malignant samples using read star
plt.plot(mal_mean, 'r*', label='malignant mean')

plt.legend(loc='upper right')
plt.title('Feature Vector Mean Values')
plt.xlabel('vector size')
plt.ylabel('mean value')
plt.show()
```

### Task 5. 20 points <a id = "section5"/>

Define a function that calculates and return the Euclidean distance between any two given data samples
 

```python
# Task 5.1 impliment the Euclidean function
def Euclidean(a, b):
    '''
        inputs:a, b: feature vectors of two patients
        
        return: d: the distance
    '''
    
    return np.sqrt(np.sum((a - b) **2))
  
# test
print('two malignant samples:', Euclidean(ftrs[1], ftrs[2])) # two maliganant samples
print('two benign samples:', Euclidean(ftrs[19], ftrs[20])) # two benign samples
print('between two mean vectors:', Euclidean(ben_mean, mal_mean)) # between means
```
